```python
import numpy as np
import scipy as sp
import pandas as pd
import sklearn
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

import sklearn.cross_validation
```

```python
# Import data
data = pd.read_csv('DS_MiniProject_ANON.csv', sep = ',')
```

```python
data
```

```python
# Rename classes for categorical data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(data.MART_STATUS)
data["MART_STATUS_T"] = le.transform(data.MART_STATUS)
le.fit(data.RTD_ST_CD)
data["RTD_ST_CD_T"] = le.transform(data.RTD_ST_CD)
le.fit(data.CustomerSegment)
data["CustomerSegment_T"] = le.transform(data.CustomerSegment)
le.fit(data.GENDER)
data["GENDER_T"] = le.transform(data.GENDER)
```

```python
data.describe()
```

```python
data.dtypes
```

```python
# Subsegment the data based on Call_Flag for exploratory data analysis
data.groupby(['Call_Flag']).describe()
```

```python
# Replace NAs with means from each category
%time
data.fillna(data.mean(), inplace=True)
```

```python
data.to_csv('data_Filled_Missing_Values.csv')
```

```python
# Response Variable
Y = np.array(data.Call_Flag)
Y
```

```python
np.sum(Y)
```

```python
#Delete non-numerical columns
data.drop('MART_STATUS', axis=1, inplace=True)
data.drop('RTD_ST_CD', axis=1, inplace=True)
data.drop('CustomerSegment', axis=1, inplace=True)
data.drop('GENDER', axis=1, inplace=True)
data.drop('DATE_FOR', axis=1, inplace=True)
data.drop('Call_Flag', axis=1, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.head()
```

```python
# Covariates
X = data.as_matrix()
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
```

```python
# 10-fold cross validation
%%time
score_arr = []
for n in xrange(1,21,1):
    clf = RandomForestClassifier(n_estimators=n, random_state=0)
    clf = clf.fit(X,Y)
    scores = cross_val_score(clf, X, Y,cv=10)
    score_arr.append(scores)
    print n
```

```python
# Plot 10-fold validation results
plt.figure(figsize=(15,8)) 
sns.boxplot(vals=score_arr)
plt.xlabel('Number of Random Forest Trees', fontweight='bold')
plt.ylabel('Accuracy (%)', fontweight='bold')
plt.title('Accuracy vs. Number of Random Forest Trees', fontweight='bold') 
plt.show()
```

```python
%%time
clf.feature_importances_
```

```python
for i in data.columns:
    print i
columnNames = np.array([i for i in data.columns])
```

```python
# Plot Feature Importance
plt.figure(figsize=(17,12)) 
sns.barplot(x = data.columns, y = clf.feature_importances_, x_order=columnNames)
plt.xlabel('Features', fontweight='bold', fontsize=28)
plt.xticks(rotation=90, fontsize=25)
plt.yticks(fontsize=25)
plt.ylabel('Feature Iportance', fontweight='bold', fontsize=28)
plt.title('Feature Importance vs. Features', fontweight='bold', fontsize=30)
plt.show()
```
